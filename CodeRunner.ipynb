```python
import os
from inference import predict

results = []

for file in os.listdir():
    if file.lower().endswith((".jpg", ".jpeg", ".png")):
        prediction, confidence = predict(file)

        gender = "Female" if prediction == 1 else "Male"
        print(f"{file} → {gender} ({confidence:.4f})")

        results.append((file, gender, confidence))
```

```python
import csv
import os
from inference import predict

# This means: look in current directory
image_folder = "."

with open("results.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Image", "Prediction", "Confidence"])

    for file in os.listdir(image_folder):
        if file.lower().endswith((".jpg", ".jpeg", ".png")):
            prediction, confidence = predict(file)
            gender = "Female" if prediction == 1 else "Male"

            print(f"{file} → {gender} ({confidence:.4f})")
            writer.writerow([file, gender, confidence])

print("✅ results.csv created successfully!")
```
